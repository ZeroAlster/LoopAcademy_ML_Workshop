### GaussianProcessClassifier

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

**Without Any Preprocessing**

```python
from sklearn.gaussian_process import GaussianProcessClassifier
gpc = GaussianProcessClassifier().fit(x_train, y_train)
gpc.score(x_test, y_test)
```

**With Preprocessing**

```python
from sklearn.pipeline import Pipeline
steps = [('scaler',StandardScaler()),
         ('Normalization',Normalizer(norm='l2')),
         ('dimension reduction', PCA(n_components=0.95)),
         ('classifier', GaussianProcessClassifier())]

clf = Pipeline(steps = steps)
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
```

###SVC Classifier

```python
from sklearn.svm import SVC, NuSVC, LinearSVC

steps = [('dimension reduction', PCA(n_components=0.95)),
         ('classifier', LinearSVC(penalty='l1', loss='squared_hinge', dual=False))]
clf = Pipeline(steps = steps)
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

###MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(100,100,100,100,100,100,100), activation='relu').fit(x_train, y_train)
mlp.score(x_test, y_test)
```
